### Serving an MLflow Model from Model Registry

**Run this in the original folder where model is registered (make sure to activate venv)**

`mlflow models serve -m "models:/model_name/version_number" -p 1234` 

port number can be anything, make sure to give proper model name and version number

```python
import requests
lst = X_test.values.tolist()
inference_request = {
        "dataframe_records": lst
}
endpoint = "http://localhost:7777/invocations"
response = requests.post(endpoint, json=inference_request)
print(response)
```
